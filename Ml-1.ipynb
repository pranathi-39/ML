```python
import numpy as np
array=np.arange(1,21)
print("array: ",array)
print("array shape: ", array.shape)
print("min and max of the array: ",array.min(),"and",max(array))
print("mean & standard deviation of the array:", array.mean(),"and",array.std())
array1=array.reshape(4,5)
print("reshaped array (4*5) : ", array1)
print("second row of the matrix: ", array1[1])
print("third row of the matrix: ",array1[2])
```

```python
import pandas as pd
data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Ravi", "Priya", "Kiran", "Anu", "Sita"],
    "Marks": [85, 92, 78, 88, 95]
}
df = pd.DataFrame(data)
print("data frame:\n", df)
print("\nFirst Three Records:")
print(df.head(3))
average = df["Marks"].mean()
print("\nAverage Marks:", average)
highest = df.loc[df["Marks"].idxmax()]
print("\nStudent with Highest Marks:",highest)
lowest = df.loc[df["Marks"].idxmin()]
print("\nStudent with Lowest Marks:",lowest)
grades = []
for mark in df["Marks"]:
    if mark >= 90:
        grades.append("A")
    elif mark >= 80:
        grades.append("B")
    else:
        grades.append("C")
df["Grade"] = grades
print("\nDataFrame with Grade:")
print(df)
```

```python
import pandas as pd
df = pd.DataFrame({
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor"],
    "Price": [50000, 500, 1200, 10000],
    "Quantity": [10, 50, 30, 15]
})
df["Total Value"] = df["Price"] * df["Quantity"]
print(df)
print("\nHighest Total Value:")
print(df.loc[df["Total Value"].idxmax()])
print("\nOverall Inventory Value:", df["Total Value"].sum())
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["Species"] = iris.target
print(df)
print(df.head(5))
print(df.shape)
print(df.columns)
print(df.isnull().sum())
print(df.describe())
print(df["Species"].value_counts())
```
