```python
#Task A: Dataset Loading and Inspection 
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "RollNo":[101,102,103,104,105,106,107,108,109,110],
    "Name":["Ravi","Priya","Kiran","Anu","Sita","Raj","Divya","Arun","Keerthi","Naveen"],
    "Gender":["M","F","M","F","F","M","F","M","F","M"],
    "Attendance":[85,92,78,95,88,70,90,82,96,75],
    "InternalMarks":[32,38,28,40,35,25,37,30,39,27],
    "ExternalMarks":[55,58,45,60,54,40,57,48,59,44]
}
df = pd.DataFrame(data)
df.to_csv("student_data.csv", index=False)
df = pd.read_csv("student_data.csv")
df
```

```python
#Task B: Statistical Analysis 
print("----- First 5 Records -----")
print(df.head())
print("\n----- Last 5 Records -----")
print(df.tail())
print("\nTotal Rows and Columns:")
print(df.shape)
print("\nColumn Names:")
print(df.columns)
print("\nData Types:")
print(df.dtypes)
print("\nDataset Information:")
df.info()
```

```python
#Task C: Data Manipulation 
print("Mean Attendance =", df["Attendance"].mean())
print("Average Internal Marks =", df["InternalMarks"].mean())
print("Average External Marks =", df["ExternalMarks"].mean())
print("Maximum Internal Marks =", df["InternalMarks"].max())
print("Minimum Internal Marks =", df["InternalMarks"].min())
print("Maximum External Marks =", df["ExternalMarks"].max())
print("Minimum External Marks =", df["ExternalMarks"].min())
print("Standard Deviation of Internal Marks =", df["InternalMarks"].std())
print("\nSummary Statistics:")
print(df.describe())
```

```python
#Task C: Data Manipulation 
df["TotalMarks"] = df["InternalMarks"] + df["ExternalMarks"]
df["Percentage"] = df["TotalMarks"]
print("----- Updated DataFrame -----")
print(df)
```

```python
#Task D: Filtering and Searching
#'1.Attendance > 90'
print("Students with Attendance greater than 90%:")
print(df[df["Attendance"] > 90])
```

```python
#'2.TotalMarks > 90'
print("Students with Total Marks greater than 90:")
print(df[df["TotalMarks"] > 90])
```

```python
#'3.Female Students'
print("Female Students:")
print(df[df["Gender"] == "F"])
```

```python
#'4.Male Students'
print("Male Students:")
print(df[df["Gender"] == "M"])
```

```python
#'5.External Marks<50'
print("Students with External Marks less than 50:")
print(df[df["ExternalMarks"] < 50])
```

```python
#'Task E: Ranking and Comparison'
#'1.Highest Total Marks'
print("Student with Highest Total Marks:")
print(df.loc[df["TotalMarks"].idxmax()])
```

```python
#'2.Lowest Total Marks'
print("Student with Lowest Total Marks:")
print(df.loc[df["TotalMarks"].idxmin()])
```

```python
#'3.Top 3 Performers'
print("Top 3 Performers:")
print(df.nlargest(3, "TotalMarks"))
```

```python
#'4.Bottom 3 Performers'
print("Bottom 3 Performers:")
print(df.nsmallest(3, "TotalMarks"))
```

```python
#'5.Above Class Average'
avg = df["TotalMarks"].mean()
print("Class Average =", avg)
print("\nStudents Scoring Above Class Average:")
print(df[df["TotalMarks"] > avg])
print("\nNumber of Students Above Class Average =", len(df[df["TotalMarks"] > avg]))
```

```python
#'Task F: Group-wise Analysis'
```

```python
#'1.Average Total Marks'
print("Average Total Marks of Male and Female Students:")
print(df.groupby("Gender")["TotalMarks"].mean())
```

```python
#'2.Average Attendance'
print("Average Attendance of Male and Female Students:")
print(df.groupby("Gender")["Attendance"].mean())
```

```python
#'Task G: Data Visualization'
```

```python
#barchart
plt.bar(df["Name"], df["TotalMarks"])
plt.title("Total Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.show()
```

```python
#Histogram
plt.hist(df["Attendance"])
plt.title("Attendance Distribution")
plt.xlabel("Attendance")
plt.ylabel("Frequency")
plt.show()
```

```python
#pie chart
gender = df["Gender"].value_counts()
plt.pie(gender, labels=gender.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
#scatter plot 
plt.scatter(df["Attendance"], df["TotalMarks"])
plt.title("Attendance vs Total Marks")
plt.xlabel("Attendance")
plt.ylabel("Total Marks")
plt.show()
print("Interpretation:")
print("Students with higher attendance generally scored higher total marks.")
```
