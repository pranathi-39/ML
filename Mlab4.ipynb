```python
import pandas as pd

# ---------------- Read Dataset ----------------
df = pd.read_csv("data.csv")

# Remove RID column if present
if "RID" in df.columns:
    df = df.drop("RID", axis=1)

target = df.columns[-1]


# ---------------- Prior Probability ----------------
def prior_probability(data):

    total = len(data)

    prior = {}

    for value in data[target]:

        prior[value] = prior.get(value, 0) + 1

    for key in prior:

        prior[key] = prior[key] / total

    return prior


# ---------------- Conditional Probability ----------------
def conditional_probability(data, attribute, attribute_value, class_value):

    class_data = data[data[target] == class_value]

    total = len(class_data)

    count = len(class_data[class_data[attribute] == attribute_value])

    if total == 0:
        return 0

    return count / total


# ---------------- Naive Bayes Prediction ----------------
def naive_bayes(data, test):

    prior = prior_probability(data)

    probability = {}

    print("\nPrior Probabilities")
    print(prior)

    print("\nCalculating Posterior Probabilities")

    for class_value in data[target].unique():

        prob = prior[class_value]

        print("\nClass :", class_value)

        for attribute in test:

            cond = conditional_probability(
                data,
                attribute,
                test[attribute],
                class_value
            )

            print(attribute, "=", test[attribute], ":", round(cond, 4))

            prob = prob * cond

        probability[class_value] = prob

        print("Posterior Probability :", round(prob, 6))

    prediction = max(probability, key=probability.get)

    return prediction


# ---------------- Sample Test Data ----------------
test_data = {
    "age": "youth",
    "income": "medium",
    "student": "yes",
    "credit_rating": "fair"
}


# ---------------- Run ----------------
result = naive_bayes(df, test_data)

print("\n==============================")
print("Prediction :", result)
print("==============================")
```
